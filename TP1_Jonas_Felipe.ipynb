Tentar criar um modelo matematico para entradas e saidas da tabela



## Bibliotecas necessárias

* Vamos importar ``tensorflow`` como ``tf``;
* ``numpy`` como ``np`` para representar os dados como listas facilmente e rapidamente;
* Vamos usar o framework ``keras`` para definir a rede neural como um conjunto de camadas seqüenciais. 


```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## Definindo e compilando a rede neural

Vamos criar o modelo mais simples de rede neural. Apenas 1 camada, essa camada tem 1 neurônio e uma entrada.

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Além do modelo, deve-se definir duas funções: a função de perda (``loss``) e otimizador (``optimizer``). 

A função de perda avalia o erro na aproximação do modelo. O ajuste é feito iterativamente e otimizador sugere os passos seguintes de modo a minimizar a função de perda.

A iterações são denominadas épocas (EPOCHS). No exemplo, utilizaremos a função de perda de erro quadrático médio ('MEAN SQUARED ERROR') e como otimizador o gradiente descendente estocástico ('STOCHASTIC GRADIENT DESCENT').


```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Fornecendo os dados


Forneceremos alguns pares (x,y) gerados com a função original  

Utilizaremos a ``numpy`` para entrada dos dados como arranjos numéricos.


Dados normalizados como sugerido 

```python
xs = np.array([0,  0.5, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([3.2, 4.1, 5.0, 6.8, 8.6, 10.4], dtype=float)
```

# Treinando a rede neural

O processo de "aprendizado" da rede neural é feito utilizando o comando ``model.fit``. Neste caso, limitou-se a 500 epochs para o ajuste. Cada época retorna o valor da função de perda avaliada. 

```python
model.fit(xs, ys, epochs=500)
```

A função ``model.predict`` pode ser usada para testar o modelo ajustado como pedido vamos testar para 38

```python
print(model.predict([3.8])*10)
```
